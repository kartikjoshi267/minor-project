```python
cnt = 0
for i in df.values:
  if i[3] in [17, 9]:
    cnt += 1

print(cnt, len(df.values))
print(cnt/len(df.values))
```

```python
from collections import Counter

cnt = 0
courses = []
for i in df.values:
  if i[1] == 14:
    cnt += 1
    courses.append(i[3])

print(cnt)
print(cnt/len(df.values))
dict(Counter(sorted(courses)))
```

```python
from collections import Counter

cnt = 0
courses = []
for i in df.values:
  if i[1] == 14 and i[34] == "Dropout":
    cnt += 1
    courses.append(i[3])

print(cnt/len(df.values))
dict(Counter(sorted(courses)))
```

```python
dfCopy = df[df["Target"] != "Enrolled"]
dfCopy = dfCopy[dfCopy["Target"] != "Graduate"]
dfCopy
```

```python
dfCopy["Target"].unique()
```

```python
for i, v in enumerate(dfCopy.columns):
  print(i, v)
```

```python
academicInclinedCols = [3,21,19,22,23,24,25,27,20]
```

```python
academicInclined = 0
notAcademicInclined = 0

for i in dfCopy.values:
  if i[9] in academicInclinedCols or i[10] in academicInclinedCols:
    academicInclined += 1
  else:
    notAcademicInclined += 1

print(academicInclined)
print(notAcademicInclined)
```
