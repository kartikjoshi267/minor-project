```python
df = df[df["Target"] == "Dropout"]
df.shape
```

```python
i = 0
for col in df.columns:
  if col != 'Target':
    fig = plt.figure(1, (10,10))
    fig = sns.displot(df[col], kind="kde", color="blue", fill=True)
    a = '/'
    fig.savefig(f"./graphs/{i}")
    i += 1
```
