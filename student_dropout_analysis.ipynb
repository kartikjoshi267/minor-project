```python
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, AdaBoostClassifier
from logitboost import LogitBoost
from xgboost import XGBClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler, StandardScaler, RobustScaler
from sklearn.impute import SimpleImputer
from sklearn.model_selection import RandomizedSearchCV
from sklearn.naive_bayes import GaussianNB
from imblearn.ensemble import BalancedRandomForestClassifier, BalancedBaggingClassifier, EasyEnsembleClassifier, RUSBoostClassifier

accuracies = {
    "SVC": [None, None],
    "Decision Tree": [None, None],
    "Random Forest": [None, None],
    "AdaBoost": [None, None],
    "LogitBoost": [None, None],
    "XGBoost": [None, None],
    "Logistic Regression": [None, None],
    "KNN": [None, None],
    "Naive Bayes": [None, None],
    "BalancedRandomForest": [None, None],
    "BalancedBaggingClassifier": [None, None],
    "EasyEnsembleClassifier": [None, None],
    "RUSBoostClassifier": [None, None],
}

classifiers = {
    "SVC": {
        "algo": SVC,
        "params": {
            "classifier__C": np.logspace(-3, 3, 7),
            "classifier__kernel": ["linear", "poly", "rbf", "sigmoid"],
            "classifier__degree": np.arange(1, 6),
            "classifier__gamma": ["scale", "auto"],
        },
    },
    "Decision Tree": {
        "algo": DecisionTreeClassifier,
        "params": {
            "classifier__criterion": ["gini", "entropy"],
            "classifier__splitter": ["best", "random"],
            "classifier__max_depth": np.arange(1, 21),
            "classifier__min_samples_split": np.arange(2, 21),
            "classifier__min_samples_leaf": np.arange(1, 21),
        }
    },
    "Random Forest": {
        "algo": RandomForestClassifier,
        "params": {
            "classifier__n_estimators": np.arange(100, 1001, 100),
            "classifier__criterion": ["gini", "entropy"],
            "classifier__max_depth": np.arange(1, 21),
            "classifier__min_samples_split": np.arange(2, 21),
            "classifier__min_samples_leaf": np.arange(1, 21),
        }
    },
    "AdaBoost": {
        "algo": AdaBoostClassifier,
        "params": {
            "classifier__n_estimators": np.arange(50, 501, 50),
            "classifier__learning_rate": np.linspace(0.01, 1, 100),
            "classifier__algorithm": ["SAMME", "SAMME.R"],
        }
    },
    "LogitBoost": {
        "algo": LogitBoost,
        "params": {
            "classifier__n_estimators": np.arange(50, 501, 50),
            "classifier__learning_rate": np.linspace(0.01, 1, 100),
        }
    },
    "XGBoost": {
        "algo": XGBClassifier,
        "params": {
            "classifier__n_estimators": np.arange(50, 101, 50),
            "classifier__learning_rate": np.linspace(0.01, 1, 100),
            "classifier__max_depth": np.arange(1, 11),
            "classifier__gamma": np.linspace(0, 1, 100),
            "classifier__subsample": np.linspace(0.1, 1, 100),
            "classifier__colsample_bytree": np.linspace(0.1, 1, 100),
        }
    },
    "Logistic Regression": {
        "algo": LogisticRegression,
        "params": {
            "classifier__C": np.logspace(-3, 3, 7),
            "classifier__solver": ["newton-cg", "lbfgs", "liblinear", "sag", "saga"],
            "classifier__max_iter": np.arange(100, 1001, 100),
        }
    },
    "KNN": {
        "algo": KNeighborsClassifier,
        "params": {
            "classifier__n_neighbors": np.arange(1, 21),
            "classifier__weights": ["uniform", "distance"],
            "classifier__algorithm": ["auto", "ball_tree", "kd_tree", "brute"],
            "classifier__p": np.arange(1, 6),
        }
    },
    "Naive Bayes": {
        "algo": GaussianNB,
        "params": {}
    },
    "BalancedRandomForest": {
        "algo": BalancedRandomForestClassifier,
        "params": {}
    },
    "BalancedBaggingClassifier": {
        "algo": BalancedBaggingClassifier,
        "params": {}
    },
    "EasyEnsembleClassifier": {
        "algo": EasyEnsembleClassifier,
        "params": {}
    },
    "RUSBoostClassifier": {
        "algo": RUSBoostClassifier,
        "params": {} 
    }
}

for key, value in classifiers.items():
    print(f"Started {key}")
    pipeline = Pipeline(steps=[
            ("imputer", SimpleImputer(strategy="mean")),
            ("scaler", RobustScaler()),
            ("classifier", value["algo"]()),
    ])

    cv = RandomizedSearchCV(pipeline, param_distributions=value["params"], cv=5, n_jobs=-1, verbose=1)
    cv.fit(X_train, y_train)
    accuracies[key][0] = cv.score(X_train, y_train).round(2)
    accuracies[key][1] = cv.score(X_test, y_test).round(2)
    accuracies[key].append(cv.best_params_)

    import pickle

    with open(f"./models pickle/{key}.pkl", "wb") as f:
        pickle.dump(cv, f)

    print(f"Ended {key}\n")
```
